```python
# Printing 1 to n in forward way
def n_nums(n:int):
    if(n==0):
        return
    n_nums(n-1)
    print(n, end=" ")

n_nums(10)
# Time complexity: O(n)
# Space complexity: O(n), cause recursive calls also n times
```
