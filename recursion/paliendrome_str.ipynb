```python
def paliendrome_str(ip_str:str, start:int, end:int):
    if(start>=end): # checking condition, for only odd start==end, 
        return True
    return (ip_str[start]==ip_str[end] and paliendrome_str(ip_str,start+1, end-1))

print(paliendrome_str("ABBA", 0, 3)) 
print(paliendrome_str("ABBC", 0, 3)) 
# Time complexity: O(n)
# Space complexity: O(n), cause every recusive calls stays on the stack untill recursion finishes. 
```
