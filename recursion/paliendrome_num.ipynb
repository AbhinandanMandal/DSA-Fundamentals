```python

def num_paliendrome(n:int):
    c_num=n
    n_num=0
    while (n>0):
        l_d=n%10
        n_num=n_num*10+l_d
        n=n//10
    if(n_num==c_num):
        return 1
    else:
        return 0
    

print(num_paliendrome(1223))
```

```python
def _paliendrome_num(n:int):
    original=n 
    last_digit=0
    while n>0:
        last_digit=last_digit*10+(n%10)
        n=n//10
    
    return original==last_digit

print(_paliendrome_num(1221))
print(_paliendrome_num(1223))

# Time complexity: O(logn)
# Space complexity: O(1)
```
