```python
def _num_factorial(n:int):
    if n==1:
        return n 
    return n*_num_factorial(n-1)

print(_num_factorial(5))
print(_num_factorial(6))
print(_num_factorial(7))
print(_num_factorial(8))
print(_num_factorial(9))
print(_num_factorial(10))

# Time complexity: O(n)
# Space complexity: O(n)
```
