```python
class Node:
    def __init__(self, key):
        self.key = key 
        self.next = None 

head = Node(10)
head.next = Node(20)
head.next.next = Node(30)

def InsertBeginning(head, key):
    temp = Node(15) # creating a node 15
    temp.next = head # connecting 15 to the head 10
    return temp # the new linked list

def PrintLinkedList(head):
    curr = head 
    while curr !=None:
        print(curr.key, end=" ")
        curr = curr.next

head = InsertBeginning(head, 15)
PrintLinkedList(head)
```
