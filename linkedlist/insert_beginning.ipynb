```python
# insert element at the beginning of the linked list

# creating a linked list
class Node:
    def __init__(self, k):
        self.key=k
        self.next=None



# approach
# let temp as node key curr
# assigning head as next of temp
# returning temp as head value
# in the solution
# initially start head as None
# head=insertbegin(head, 10), so head=None, temp=Node(10), temp.next=head=None, returns temp=10, so head=10
# head=isnertbegin(head,20), temp=Node(20), temp.next=head = temp.next=10, so it becomes 20->10->None, and new head=20


def insertbegin(head, key):
    temp=Node(key) 
    temp.next=head
    return temp

head=None
head=insertbegin(head, 10) # head assigned to 10, 10->None
head=insertbegin(head,20) # head assigned to 20, next is 10, 20->10->None
head=insertbegin(head, 30) # 30->20->10->None   

# time complexity: O(1)
# space complexity: O(1)
```

```python
# printing linkedlist
def print_linkedlist(head):
    curr=head
    while curr!=None:
        print(curr.key, end=" ")
        curr=curr.next

print_linkedlist(head=head)
```
