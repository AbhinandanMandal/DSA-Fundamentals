```python
class Node:
    def __init__(self, k):
        self.key = k
        self.next = None


head = Node(10)
head.next = Node(20)
head.next.next = Node(30)

def LengthLinkedList(head):
    curr = head 
    count=0
    while curr!=None:
        count+=1
        curr = curr.next 
    print(count)

LengthLinkedList(head)
```
