```python
# making a linked list
class Node:
    def __init__(self, k):
        self.key=k
        self.next=None


head=Node(10)
head.next=Node(20)
head.next.next=Node(15)
head.next.next.next=Node(15)

# printing content of linekd list
def printlikedlist(head):
    curr=head
    while curr!=None:
        print(curr.key, end=" ")
        curr=curr.next



printlikedlist(head=head)
# time complexity: O(n)
```
