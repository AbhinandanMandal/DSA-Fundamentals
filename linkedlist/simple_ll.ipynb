```python
# implementation of simple linked list
class Node:
    def __init__(self, k):
        self.key=k
        self.next=None

temp1=Node(10)
temp2=Node(20)
temp3=Node(30)

# forming of linked list 
temp1.next=temp2
temp2.next=temp3
head=temp1

print(head.key)
print(temp2.key)
print(temp3.key)
```

```python
# another easy way to do this
class Node:
    def __init__(self, k):
        self.key = k
        self.next = None

# creating linked list
head=Node(10)
head.next=Node(20)
head.next.next=Node(30)

print(head.key)
print(head.next.key)
print(head.next.next.key)
```
