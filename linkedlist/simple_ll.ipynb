```python
# length of a ll
# making a linked list
class Node:
    def __init__(self, k):  # k is key in linked list
        self.key = k
        self.next = None


head = Node(10)  # head of the linked list
head.next = Node(20)  # selecting next nodes cronologically
head.next.next = Node(40)


def len_ll(head):
    curr=head
    count=0
    while curr!=None:
        count+=1
        curr=curr.next
    return count

print(len_ll(head=head))
```

```python
# searcing ll
# making a linked list
class Node:
    def __init__(self, k):  # k is key in linked list
        self.key = k
        self.next = None


head = Node(10)  # head of the linked list
head.next = Node(20)  # selecting next nodes cronologically
head.next.next = Node(30)
head.next.next.next = Node(40)
head.next.next.next.next = Node(50)

def search_ll(head, n):
    curr=head
    pos=0
    while curr!=None:
        pos+=1
        if curr.key==n:
            return pos
        curr=curr.next


print(search_ll(head=head, n=30))

```

```python
# insert at the beginning
# making a linked list
class Node:
    def __init__(self, k):  # k is key in linked list
        self.key = k
        self.next = None


head = Node(10)  # head of the linked list
head.next = Node(20)  # selecting next nodes cronologically
head.next.next = Node(30)
head.next.next.next = Node(40)
head.next.next.next.next = Node(50)


def insert_beginning(head,n):
    temp=Node(n)
    temp.next=head
    return temp # returning new head is compulsary to make updated ll

head=head
head=insert_beginning(head, 5)


def print_ll(head):
    curr = head
    while curr != None:
        print(curr.key, end=" ")
        curr = curr.next


print_ll(head=head)
```

```python
# inserting at end
class Node:
    def __init__(self, k):  # k is key in linked list
        self.key = k
        self.next = None


head = Node(10)  # head of the linked list
head.next = Node(20)  # selecting next nodes cronologically
head.next.next = Node(30)
head.next.next.next = Node(40)
head.next.next.next.next = Node(50)



def insert_ending(head, n):

    # if head is None
    # add node and made  ll (n->None)
    if head==None:
        return Node(n) 
    
    curr=head
    while curr.next!=None: # traverse till curr is None value
        curr=curr.next
    
    # when curr becomes None
    # append new node after curr
    curr.next=Node(n)
    return head

head=head
head=insert_ending(head,60)



def print_ll(head):
    curr = head
    while curr!= None:
        print(curr.key, end=" ")
        curr = curr.next


print_ll(head=head)
```

```python
# insert at particular position
class Node:
    def __init__(self, k):  # k is key in linked list
        self.key = k
        self.next = None


head = Node(10)  # head of the linked list
head.next = Node(20)  # selecting next nodes cronologically
head.next.next = Node(30)
head.next.next.next = Node(40)
head.next.next.next.next = Node(50)



def insert_position(head, key, pos):
    curr=head
    count=1
    temp=Node(key)

    if pos==1: # if position is 1
        temp.next=head
        return head

    while count<pos-1: # this is important like to actually insert positon
        curr=curr.next
        count+=1
        if curr==None:
            return head

    # this is the core of inserting in a particular position
    temp.next=curr.next
    curr.next=temp
    return head
    
    
head=head
head=insert_position(head,45,4)


def print_ll(head):
    curr = head
    while curr != None:
        print(curr.key, end=" ")
        curr = curr.next


print_ll(head=head)

```
