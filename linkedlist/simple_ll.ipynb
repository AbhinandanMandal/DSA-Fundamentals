```python
# Shortcut Method
class Node:
    def __init__(self, k):
        self.key = k
        self.next=None 

head=Node(10)
head.next=Node(20)
head.next.next = Node(30)

print(head.key)
print(head.next.key)
print(head.next.next.key)
```

```python
# making a linked list
class Node:
    def __init__(self, k): # k is key in linked list
        self.key=k
        self.next=None

head=Node(10) # head of the linked list
head.next=Node(20) # selecting next nodes cronologically
head.next.next=Node(40)



# print(head.key)
# print(head.next.key)
# print(head.next.next.key)

# traversing linked list
# another method for printing the ll
def print_ll(head):
    curr=head
    while curr!=None:
        print(curr.key)
        curr=curr.next

print_ll(head=head)
```

```python
# length of a ll
# making a linked list
class Node:
    def __init__(self, k):  # k is key in linked list
        self.key = k
        self.next = None


head = Node(10)  # head of the linked list
head.next = Node(20)  # selecting next nodes cronologically
head.next.next = Node(40)


def len_ll(head):
    curr=head
    count=0
    while curr!=None:
        count+=1
        curr=curr.next
    return count

print(len_ll(head=head))
```
