```python
class Node:
    def __init__(self, key):
        self.key = key 
        self.next = None 

head = Node(10)
head.next = Node(20)
head.next.next = Node(30)

def InsertEndPosition(head, key):
    temp = Node(key)
    if head is None:
        return temp
    curr = head
    while curr.next!=None: # remeber while loop
        curr = curr.next
    curr.next = temp
    return head 

def PrintLinkedList(head):
    curr = head
    while curr!=None:
        print(curr.key, end=" ")
        curr = curr.next

head = InsertEndPosition(head, 15)
PrintLinkedList(head)
```
