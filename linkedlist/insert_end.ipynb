```python
# creating a linked list
class Node:
    def __init__(self,k):
        self.key=k
        self.next=None

# time complexity: o(1)
# insert at the beginning
def insert_begin(head, key):
    curr=Node(key)
    curr.next=head # shifting head one by one
    return curr

head=None
head=insert_begin(head,10)
head=insert_begin(head, 20)

# printing linked list
def print_linkedlist(head):
    curr = head
    while curr != None:
        print(curr.key, end=" ")
        curr = curr.next


print_linkedlist(head=head)
```

```python
# creating a linked list
class Node:
    def __init__(self, k):
        self.key = k
        self.next = None



# insert element at end
# time complexity: O(n)
def insert_end(head, key):
    if head==None:
        return Node(key)
    curr=head
    while head.next!=None:
        curr=curr.next
    curr.next=Node(key)
    return head 


head = None
head = insert_end(head, 10)
head = insert_end(head, 20)

# printing linked list
def print_linkedlist(head):
    curr = head
    while curr != None:
        print(curr.key, end=" ")
        curr = curr.next


print_linkedlist(head=head)
```

```python
# linked list
class Node:
    def __init__(self, k):
        self.key = k
        self.next = None

# insert end
def insert_end(head, key):
    
    # for initial condition when head is none, return the key value as head
    if head==None:
        return Node(key)
    
    # at 2nd iteration, assign curr as head and running a while loop, is curr.next not equals to None 
    # then inc curr and when it's none then make it node (key)
    curr=head
    while curr.next!=None:
        curr=curr.next
    curr.next=Node(key)
    return head


head=None
head=insert_end(head,10)
head=insert_end(head,20)
head=insert_end(head,30)

# printing linked list
def print_linkedlist(head):
    curr = head
    while curr != None:
        print(curr.key, end=" ")
        curr = curr.next


print_linkedlist(head=head)
 
```
