```python
# creating a linked list
class Node:
    def __init__(self,k):
        self.key=k
        self.next=None

# time complexity: o(1)
# insert at the beginning
def insert_begin(head, key):
    curr=Node(key)
    curr.next=head # shifting head one by one
    return curr

head=None
head=insert_begin(head,10)
head=insert_begin(head, 20)

# printing linked list
def print_linkedlist(head):
    curr = head
    while curr != None:
        print(curr.key, end=" ")
        curr = curr.next


print_linkedlist(head=head)
```

```python
# creating a linked list
class Node:
    def __init__(self, k):
        self.key = k
        self.next = None



# insert element at end
# time complexity: O(n)
def insert_end(head, key):
    if head==None:
        return Node(key)
    curr=head
    while head.next!=None:
        curr=curr.next
    curr.next=Node(key)
    return head 


head = None
head = insert_end(head, 10)
head = insert_end(head, 20)

# printing linked list
def print_linkedlist(head):
    curr = head
    while curr != None:
        print(curr.key, end=" ")
        curr = curr.next


print_linkedlist(head=head)
```
