```python
# making a linked list
class Node:
    def __init__(self, k):  # k is key in linked list
        self.key = k
        self.next = None


head = Node(10)  # head of the linked list
head.next = Node(20)  # selecting next nodes cronologically
head.next.next = Node(30)
head.next.next.next = Node(40)
head.next.next.next.next = Node(50)

# head=None

def del_lastnode(head):

    if head==None: # if the 1st node is none itself
        return None

    if head.next==None: # if 10 -> none scenario
        return None
    
    curr=head # normal scenario
    while curr.next.next!=None: # seeing far sight 10-20-30-none, so seeing 30 from 10
        curr=curr.next
    curr.next=None
    return head

head=del_lastnode(head=head)


def print_ll(head):
    curr = head
    while curr != None:
        print(curr.key, end=" ")
        curr = curr.next

print_ll(head=head)

```

```python
class Node:
    def __init__(self, key):
        self.key = key
        self.next = None


head = Node(10)
head.next = Node(20)
head.next.next = Node(30)

def DelLastNode(head):
    if head == None:
        return None 
    
    if head.next == None:
        return None 
    
    curr = head 
    while curr.next.next != None:
        curr = curr.next
    curr.next = None
    return head 


def PrintLinkedList(head):
    curr = head
    while curr != None:
        print(curr.key, end=" ")
        curr = curr.next


head = DelLastNode(head)
PrintLinkedList(head)
```
