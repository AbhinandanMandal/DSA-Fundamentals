```python
# while loop
# linked list
class Node:
    def __init__(self, k):
        self.key = k
        self.next = None


head = Node(10)
head.next = Node(30)
head.next.next = Node(50)
head.next.next.next = Node(70)



# insert at specific position
def insert_pos(head, pos, val):
    temp = Node(val)
    if pos == 1:
        temp.next = head
        return head

    curr = head
    count = 1
    while count < pos-1:
        curr = curr.next
        count += 1
        if curr == None:
            return head

    temp.next = curr.next
    curr.next = temp
    return head

head = insert_pos(head=head, pos=2, val=20)


# printing linked list
def printlikedlist(head):
    curr = head
    while curr != None:
        print(curr.key, end=" ")
        curr = curr.next


printlikedlist(head=head)
```

```python
# insert at a given position
# linked list
class Node:
    def __init__(self, k):
        self.key = k
        self.next = None


head = Node(10)
head.next = Node(30)
head.next.next = Node(50)
head.next.next.next = Node(70)


# insert at specific position
def insert_pos(head,pos, val):

    # storing value in new_node
    new_node=Node(val)

    # if position becomes 1 then new_node places at the position 1 and others become the next nodes
    if pos==1:
        new_node.next=head
        return new_node
    
    # if position is not 1 then
    # store head into curr and count=1
    # running a while loop to check if curr is not none and current position should be less than pos-1
    # eg. 1 - 2 - 3, and position given as 2 
    # so, we need to go till 1 (2-1) then break the node, insert new node and join it

    curr=head
    count=1
    while curr is not None and count<pos-1:
        curr=curr.next # update the curr and count
        count+=1
    
    # breaking and joining
    # initially new_node=Node(val)
    # so, next values of new_node will the values of curr after specified node
    # and before new_node all the values of current places
    new_node.next=curr.next
    curr.next=new_node
    return head

head=insert_pos(head,3,20)


def printlikedlist(head):
    curr = head
    while curr != None:
        print(curr.key, end=" ")
        curr = curr.next

printlikedlist(head=head)

# time complexity: O(n)
```
