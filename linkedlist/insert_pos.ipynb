```python
class Node:
    def __init__(self, key):
        self.key = key 
        self.next = None 


head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
head.next.next.next = Node(40)

def InsertPosition(head, key, pos):
    temp  = Node(key)
    curr = head

    if pos == 1:
        temp.next = head 
        return temp 
    
    for _ in range(pos - 2):
        curr = curr.next
    temp.next = curr.next
    curr.next = temp
    return head

def PrintLinkedList(head):
    curr = head 
    while curr!=None:
        print(curr.key, end=" ")
        curr = curr.next 

head = InsertPosition(head, 15, 4)
PrintLinkedList(head)
```
