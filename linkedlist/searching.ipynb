```python
class Node:
    def __init__(self, k):
        self.key = k
        self.next = None 

head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
head.next.next.next = Node(40)
head.next.next.next.next = Node(50)

def SearchLinkedList(head, n):
    pos = 0
    curr = head 
    while curr != None:
        if curr.key == n:
            return pos 
        pos+=1
        curr=curr.next 

print(SearchLinkedList(head, 30))
```
