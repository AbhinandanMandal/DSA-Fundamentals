```python
# searching in a linked list
# making a linked list
class Node:
    def __init__(self, k):
        self.key = k
        self.next = None


head = Node(10)
head.next = Node(5)
head.next.next = Node(20)
head.next.next.next = Node(15)


def search_linkedlist(head, n):
    curr = head
    count = 0
    while curr != None:
        count += 1
        if curr.key == n: # if curr item is n then return item's position
            return count
        curr=curr.next # getting next item 

    return -1

print(search_linkedlist(head=head,n=20))
```
