```python
# delete a node with head
# making a linked list
class Node:
    def __init__(self, k):  # k is key in linked list
        self.key = k
        self.next = None


head = Node(10)  # head of the linked list
head.next = Node(20)  # selecting next nodes cronologically
head.next.next = Node(30)
head.next.next.next = Node(40)
head.next.next.next.next = Node(50)

def del_node_head(head,n): # delete a node with given head
    curr=head
    while curr.next!=None and curr.next.key!=n: # core 1
        curr=curr.next
        
    if curr.next!=None:
        curr.next=curr.next.next # core 2

    return head

head=del_node_head(head=head, n=30)


def print_ll(head):
    curr = head
    while curr != None:
        print(curr.key, end=" ")
        curr = curr.next

print_ll(head=head)

```

```python
# delete a node without head
# making a linked list
class Node:
    def __init__(self, k):  # k is key in linked list
        self.key = k
        self.next = None


head = Node(10)  # head of the linked list
head.next = Node(20)  # selecting next nodes cronologically
head.next.next = Node(30)
head.next.next.next = Node(40)
head.next.next.next.next = Node(50)


def delnode_wohead(pointer):
    pointer.key=pointer.next.key # 30 becomes 40, 10,20,40,40,50
    pointer.next=pointer.next.next # 10,20,40,50
    

pointer=head.next.next # 30
delnode_wohead(pointer=pointer) # deleting node 30


def print_ll(head): # printing the updated ll
    curr = head
    while curr != None:
        print(curr.key, end=" ")
        curr = curr.next
print_ll(head=head)


```
