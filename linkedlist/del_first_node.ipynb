```python
# linked list
class Node:
    def __init__(self, k):
        self.key = k
        self.next = None

head=Node(10)
head.next=Node(20)
head.next.next=Node(30)
head.next.next.next=Node(40)


# delete first node
def del_first_node(head):
    
    if head==None:
        return None
    return head.next

head=del_first_node(head=head)

# printing linked list
def print_ll(head):
    curr = head
    while curr != None:
        print(curr.key, end=" ")
        curr = curr.next
        
print_ll(head=head)
```
