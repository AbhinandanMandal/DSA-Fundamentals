```python

def isSorted(nums:list):
    incFlag=True
    decFlag=True
    n=len(nums)
    if(n<=1):
        return 1
    else:
        for i in range(n-1):
            if (nums[i]>nums[i+1]):
                incFlag=False
            if (nums[i]<nums[i+1]):
                decFlag=False
            if not incFlag and not decFlag:
                return 0
    return 1


print(isSorted([10, 20, 30]))
print(isSorted([30, 20, 10]))
print(isSorted([30, 10,20]))
```

```python
def _is_array_sorted(nums:list)->int:
    incFlag=True
    decFlag=True

    if len(nums)<=1:
        return 1 # 1 is sorted, 0 is not sorted
    
    for i in range(len(nums)-1):
        if (nums[i]>nums[i+1]):
            incFlag=False
        if (nums[i]<nums[i+1]):
            decFlag=False
        if not incFlag and not decFlag:
            return 0
    return 1


print(_is_array_sorted([10, 20, 30]))
print(_is_array_sorted([30, 20, 10]))
print(_is_array_sorted([30, 10, 20]))

# Time complexity: O(n)
# Space complexity: O(1)
```
