```python
def list_reverse(nums:list):
    if(len(nums)<=1):
        return nums
    else:
        nums=nums[::-1]
    return nums


nums = [1, 2, 3, 4, 5]
print(list_reverse(nums=nums)) 
# Time complexity: O(n)
# Space complexity: O(1)
```

```python
def list_reverse(nums:list):
    n=len(nums)
    s=0
    e=n-1
    while s<e:
        nums[s],nums[e]=nums[e],nums[s]
        s+=1
        e-=1
    return nums


nums = [1, 2, 3, 4, 5]
print(list_reverse(nums=nums)) 
# Time complexity: O(n)
# Space complexity: O(1)
```

```python
def _reverse_list(nums:list):
    return nums[::-1]

print(_reverse_list([1,2,3,4,5]))
print(_reverse_list([1,2]))
print(_reverse_list([1]))

# Time complexity: O(n)
# Space complexity: O(1)
```
