```python
# method 1
def largest_element(nums:list):
    max=nums[0]
    if not nums:
        return None
    for i in range(1,len(nums)):
        if nums[i]>max:
            max=nums[i]
    return max


nums = [10, 5, 20, 8]
print(largest_element(nums=nums))  
# Time complexity: O(n)
# Space complexity: O(1)
```

```python
def _largest_element(nums:list)->int:
    element=nums[0]
    if not nums:
        return None 
    for i in nums:
        if i>element:
            element=i
    return element


nums = [10, 5, 20, 8]
print(_largest_element(nums=nums))  
# Time complexity: O(n)
# Space complexity: O(1)
```
