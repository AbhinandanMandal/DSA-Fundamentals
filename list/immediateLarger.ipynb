```python
def _get_immediate_larget(nums:list, n:int)->int:
    ans=-1 
    diff = 100000
    for i in nums:
        if i>n and i-n<diff:
            ans=i
            diff=i-n
    return ans 


nums = [63, 6, 60, 38, 3, 94, 93, 83, 65]
n = 9
print(_get_immediate_larget(nums=nums, n=n))
# Time complexity: O(n)
# Space complexity: O(1)
```
