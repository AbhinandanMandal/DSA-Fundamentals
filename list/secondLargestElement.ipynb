```python
def second_largest(nums:list):
    max_1=max(nums)
    max_2=None
    n=len(nums)
    if (n<=1):
        return None
    else:
        for i in nums:
            if (i!=max_1):
                if max_2==None:
                    max_2=i
                else:
                    max_2=max(i, max_2)
    return max_2
    

nums = [1, 6, 3, 4, 5]
print(second_largest(nums=nums))  
# Time complexity: O(n)
# Space complexity : O(1)
```

```python
def _get_second_largest(nums:list):
    if len(nums)<2:
        return None
    
    max1 = max2 = float('-inf')
    for i in nums:
        if i>max1:
            max2=max1
            max1=i 
        elif max1>i>max2:
            max2=i 
    return None if max2 == float('-inf') else max2


print(_get_second_largest([1, 6, 3, 4, 5]))   # 5
print(_get_second_largest([10, 10, 5]))       # 5
print(_get_second_largest([7]))               # None

# Time complexity: O(n)
# Space complexity: O(1)
```
