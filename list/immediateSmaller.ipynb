```python
def _get_immediate_smaller(nums:list, n:int)->int:
    ans=-1
    diff = 1000000
    for i in nums:
        if (i<n and n-i <diff):
            ans=i
            diff=n-i
    return ans 


nums = [63, 6, 60, 38, 3, 94, 93, 83, 65]
x = 9
print(_get_immediate_smaller(nums, x))
# Time complexity: O(n)
# Space complexity: O(1)
```
