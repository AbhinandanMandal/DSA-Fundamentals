```python
def _left_rotate(nums:list):
    n1=nums[0]
    for i in range(1, len(nums)):
        nums[i-1]=nums[i]
    nums[len(nums)-1]=n1
    return nums

print(_left_rotate([1,2,3,4]))
print(_left_rotate([1,2,3,4,5,6]))
print(_left_rotate([1,2,3,4,5,6,7,8]))

# Time complexity: O(n)
# Space complexity: O(1)
```
