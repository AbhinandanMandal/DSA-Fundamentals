Bubble sort: 
simplest sorting algorithm, it's `stable` and `in-place`.

Basic of Bubble sort is: it chooses the large value and put it in the end position of the array and keep doing it 
                         untill it become sorted array.

```python
def _bubble_sort(arr:list):
    for i in range(len(arr)-1):
        for j in range(len(arr)-i-1):
            if arr[j]>arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr 

arr=[10, 8, 20, 5]
print(_bubble_sort(arr))

# Time complexity: O(n^2)
# Space complexity: O(1)
```

```python

# If array is sorted from the middle then
def bubble_sort(nums:list):
    sorted_nums=sorted(nums)
    for i in range(len(nums)-1):
        for j in range(len(nums)-i-1):
            if (nums[j] > nums[j+1]):
                nums[j], nums[j+1] = nums[j+1], nums[j]
        if (nums==sorted_nums):
            return nums
    return nums


nums = [8, 5, 2, 7, 9, 10]
print(bubble_sort(nums=nums))

# Time complexity: O(n^2)
# Space complexity: O(1)
```

```python
def bubble_sort(nums:list):
    for i in range(len(nums)-1):
        swapped=False
        for j in range(len(nums)-i-1):
            if (nums[j] > nums[j+1]):
                nums[j], nums[j+1] = nums[j+1], nums[j]
                swapped=True
        if swapped==False:
            return nums
    return nums


nums = [8, 5, 2, 7, 9, 10]
print(bubble_sort(nums=nums))

# Time complexity: O(n^2)
# Space complexity: O(1)
```
