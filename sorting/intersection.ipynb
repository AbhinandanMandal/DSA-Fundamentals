```python
# worst method
def intersection_array(nums1:list, nums2:list):
    m=len(nums1)
    n=len(nums2)
    result=set()
    if(m>n):
        for i in range(m):
            if(nums1[i] in nums2):
                result.add(nums1[i])
    else:
        for i in range(n):
            if(nums2[i] in nums1):
                result.add(nums2[i])
    return list(result)


nums1 = [1, 1, 3, 3]
nums2 = [1, 1, 3, 5, 7]
print(intersection_array(nums1=nums1, nums2=nums2))

# Time complexity: O(m.n)
# Space complexity: O(min(m,n))
```

```python
# optimized method
def _intersection_array(arr1:list, arr2:list):
    intersection=set()
    arr1=set(arr1)
    for i in arr2:
        if i in arr1:
            intersection.add(i)
    return list(intersection)


nums1 = [1, 1, 3, 3]
nums2 = [1, 1, 3, 5, 7]
print(_intersection_array(nums1, nums2))

# Time complexity: O(m+n)
# Space complexity: O(m+n)
```

```python
# a more optimized method without using inbuilt set()
def _intersection_array(arr1:list, arr2:list):
    n=len(arr1)
    m=len(arr2)
    i=0
    j=0
    result=[]
    while(i<n and j<m):

        if(i>0 and arr1[i]==arr2[i-1]): # we're thinking on prospect of i, also thinking in prospect of j can be done
            i+=1
            continue

        if(arr1[i]==arr2[j]): # start understanding from this line
            result.append(arr1[i])
            i+=1
            j+=1

        if(arr1[i]<arr2[j]):
            i+=1
        else:
            j+=1

    return result


nums1 = [1, 1, 3, 3]
nums2 = [1, 1, 3, 5, 7]
print(_intersection_array(nums1, nums2))

# Time complexity: O(m+n)
# Space complexity: O(m+n)
```
