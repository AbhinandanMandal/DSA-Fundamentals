Merge subarray can be done by bifurcating array into two, then apply merge two sorted array concept

```python
def _merge_subarray(arr:list, low:int, high:int, mid:int):
    arr1=arr[low:mid+1]
    arr2=arr[mid+1:high+1]
    n=len(arr1)
    m=len(arr2)
    i=0
    j=0
    result=[]

    while(i<n and j<m):
        if (arr1[i]<arr2[j]):
            result.append(arr1[i])
            i+=1
        else:
            result.append(arr2[j])
            j+=1
    while(i<n):
        result.append(arr1[i])
        i+=1
    while(j<m):
        result.append(arr2[j])
        j+=1
    return result

arr=[5,8,11,14,7]
print(_merge_subarray(arr,0, 4, 3))

# Time complexity: O(m+n)
# Space complexity: O(m+n)
```
