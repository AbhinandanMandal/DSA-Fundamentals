**partition**

Let say in an unsorted array [5,4,1,2,3,6,7], we want all the element on left side that are less than 3 and all the element on right side that is larger than 3. 

If any element equals to 3 (3) then that also have to exist on the left side of 3.

So, after partition, list going to be [1,2,3,4,5,6,7]

```python
# Naive approach
def naive_partition(arr:list, p:int):
    n=len(arr)
    arr[p],arr[n-1]=arr[n-1],arr[p] # swapping the current pth element with the last element
    result=[]

    for i in range(n):
        if arr[i]<=arr[n-1]:
            result.append(arr[i])
    for j in range(n):
        if arr[j]>arr[n-1]:
            result.append(arr[j])
    return result

arr=[5, 13,6,9,12,8,11]
print(naive_partition(arr, 3))

# Time complexity: O(n)
# Space complexity: O(n)
```

```python
# method 2
def lomuto_parition(nums:list, p:int):
    """p:partition number index """
    n=len(nums)
    nums[p],nums[n-1]=nums[n-1],nums[p]
    pivot=nums[n-1]
    i=-1
    for j in range(n-1): # excluding pivot element from operation
        if(nums[j]<pivot):
            i+=1
            nums[i],nums[j]=nums[j],nums[i]
    nums[i+1],nums[n-1]=nums[n-1],nums[i+1]
    return nums
```

```python
nums=[5, 13,6,9,12,8,11]
print(lomuto_parition(nums=nums,p=3))
# time complexity: O(n)
# space complexity: O(1)
```

```python
# method 3
def hoare_partition(nums:list, low:int, high:int, p:int):
    """p: index of pivot number"""
    pivot=nums[p]
    i=low-1
    j=high+1
    while True:
        i+=1
        while(nums[i]<pivot):
            i+=1
        j-=1
        while(nums[j]>pivot):
            j-=1
        if(i>=j):
            return j # returns the index number of pivot after sucessfull partition
        nums[i],nums[j]=nums[j],nums[i]

```

```python
nums=[5, 13,6,9,12,8,11]
print(hoare_partition(nums=nums, low=0, high=6, p=2))
# time complexity: O(n)
# space complexity: O(1)
```
