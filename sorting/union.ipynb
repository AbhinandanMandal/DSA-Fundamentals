```python
# if if loop using
def union_array(a:list, b:list):
    res=[]
    m=len(a)
    n=len(b)
    i=0
    j=0
    while(i<m and j<n):
        if (a[i]<=b[j]):
            if(a[i] not in res):
                res.append(a[i])
            i+=1
        else:
            if (b[j] not in res):
                res.append(b[j])
            j+=1

    while(i<m):
        if (a[i] not in res):
            res.append(a[i])
        i+=1

    while(j<n):
        if(b[j] not in res):
            res.append(b[j])
        j+=1

    return res


a = [2, 20, 20, 40]
b = [1, 10, 20]
print(union_array(a, b))

# Time complexity: O(m+n)
# Space complexity: O(m+n)
```

```python
# Baseline method using for loop
def _union_of_array(arr1:list, arr2:list):
    union=[]
    for i in arr1:
        if i not in union:
            union.append(i)
    for j in arr2:
        if j not in union:
            union.append(j)
    return union


a = [2, 20, 20, 40]
b = [1, 10, 20]
print(_union_of_array(a,b))

# Time complexity: O(m+n)
# Space complexity: O(m+n)
```

```python
# Concept of merge subarray can be utilized
def _union_of_array(arr1:list, arr2:list):
    n=len(arr1)
    m=len(arr2)
    i=0
    j=0
    union=[]
    while(i<n and j<m):
        if(arr1[i]<=arr2[j]):
            union.append(arr1[i])
            i+=1
        else:
            union.append(arr2[j])
            j+=1
    while(i<n):
        union.append(arr1[i])
        i+=1
    while(j<m):
        union.append(arr2[j])
        j+=1

    return list(set(union))


a = [2, 20, 20, 40]
b = [1, 10, 20]
print(_union_of_array(a, b))

# Time complexity: O(m+n)
# Space complexity: O(m+n)
```
