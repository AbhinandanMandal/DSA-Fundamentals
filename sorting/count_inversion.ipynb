**what is count inversion**

An inversion count in an array measures how far or close an array is from being sorted. It represents the total number of pairs of elements a[i],a[j] such that (i<j) but (a[i]>a[j]). 

A sorted array has 0 inversions, while a reverse-sorted array has the maximum number of inversions. 

```python
# Baseline method
def count_inversion(arr:list):
    inversion=0
    n=len(arr)

    # If completely different
    if(arr == sorted(arr, reverse=True)):
        return (n*(n-1))//2
    
    else:
        for i in range(n):
            for j in range(i+1,n):
                if (arr[i]>arr[j]):
                    inversion+=1
    return inversion
    
array1 = [2, 4,1,3,5]
array2 = [5,4,3,2,1]
print(count_inversion(array1))
print(count_inversion(array2))

# Time complexity: O(n.n + nlogn), nlogn for sorted reverse
# Space complexity: O(n)
```

```python
# using modified sorted array
def count_merge_subarray(arr:list, low:int, high:int, mid:int):
    arr1=arr[low:mid+1]
    arr2=arr[mid+1:high+1]
    n=len(arr1)
    m=len(arr2)
    i=0
    j=0
    k=low 
    count=0 # for inversion
    while(i<n and j<m):
        if (arr1[i]<arr2[j]):
            arr[k]=arr1[i]
            i+=1
        else:
            arr[k]=arr2[j]
            j+=1
            count+=(m-i) # we'll only considering inversion of arr1[i]>arr2[j]
        k+=1
    while(i<n):
        arr[k]+=arr1[i]
        i+=1
        k+=1
    while(j<m):
        arr[k]+=arr2[j]
        j+=1
        k+=1
    return count 


def count_inversion(arr:list, low:int, high:int):
    count=0
    if (high>low):
        mid=(low+high)//2
        count+=count_inversion(arr, low, mid)
        count+=count_inversion(arr, mid+1, high)
        count+=count_merge_subarray(arr, low, high, mid)
    return count


nums = [5, 8, 11, 14, 2, 7]
print(count_inversion(arr=nums, low=0, high=len(nums)-1))
```
