**lomuto partition**

```python
def lomuto_partition(arr:list, p:int):
    n=len(arr)
    pivot=arr[p]
    arr[p],arr[n-1]=arr[n-1],arr[p] # swapping
    i=-1

    for j in range(n-1):
        if arr[j]<pivot:
            i+=1
            arr[i],arr[j]=arr[j],arr[i]
    arr[i+1],arr[n-1]=arr[n-1],arr[i+1]
    return arr

arr = [5, 13,6,9,12,8,11]
print(lomuto_partition(arr, 3))

# Time complexity: O(n)
# Space complexity: O(1)
```
