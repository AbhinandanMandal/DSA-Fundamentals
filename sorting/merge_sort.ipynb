Merge Sort: stable, not-in place algorithm

```python
def _merge_subarray(arr:list, low:int, high:int, mid:int):
    arr1=arr[low:mid+1]
    arr2=arr[mid+1:high+1]
    i=0
    j=0
    n=len(arr1)
    m=len(arr2)
    k=low
    while(i<n and j<m):
        if (arr1[i]<=arr2[j]):
            arr[k]=arr1[i]
            i+=1
            k+=1
        else:
            arr[k]=arr2[j]
            j+=1
            k+=1
    while(i<n):
        arr[k]=arr1[i]
        i+=1
        k+=1
    while (j<m):
        arr[k]=arr2[j]
        j+=1
        k+=1
    
    return arr

def merge_sort(arr:list, low:int, high:int):
    """ 
    arr: array
    low: first index -> 0
    high: last index -> len(arr)-1
    """
    if (high>low):
        mid=(low+high)//2
        merge_sort(arr, low, mid)
        merge_sort(arr, mid+1, high)
        _merge_subarray(arr, low, high, mid)
    return arr

arr = [10, 5, 30, 15, 7]
print(merge_sort(arr, 0, 4))

# Time complexity: O(nlogn)
# Space complexity: O(n)
```
