```python
def _merge_two_sorted_arrays(arr1:list, arr2:list):
    n=len(arr1)
    m=len(arr2)
    i=0
    j=0
    result=[] # It'll take an extra space for this
    while (i<n and j<m):
        if (arr1[i]<arr2[j]):
            result.append(arr1[i])
            i+=1
        else:
            result.append(arr2[j])
            j+=1
    while(i<n):
        result.append(arr1[i])
        i+=1
    while(j<m):
        result.append(arr2[j])
        j+=1
    return result

arr1=[10, 15]
arr2=[5,6,6,30,40]
print(_merge_two_sorted_arrays(arr1, arr2))

# Time complexity: O(m+n)
# space complexity: O(m+n)
```
