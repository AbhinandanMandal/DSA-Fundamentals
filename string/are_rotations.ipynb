```python
def areRotations(s1:str, s2:str):
    if len(s1)!=len(s2):
        return False
    temp=s1+s1 # concating s1 string
    return temp.find(s2)!=-1 # returning the index of s2 in the concate string else -1

    

s1='ABCD'
s2='CDAA'
print(areRotations(s1=s1,s2=s2))
# time complexity: O(n)
```
