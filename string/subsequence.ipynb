```python
# printing subsquence of a string
def subsequence(s:str, output):
    if len(s)==0:
        print(output, end=' ')
        return
    subsequence(s[1:],output+s[0])
    subsequence(s[1:],output)

subsequence(s='abc', output='')
```

```python
def is_subsequence(s1:str, s2:str):
    i=0
    j=0
    while(i<len(s1) and j<len(s2)):
        if s1[i]==s2[j]:
            j+=1
        i+=1 # increasing i in every step

    if j==len(s2):
        return True
    else:
        return False

s1='ABCDE'
s2='ADE'

print(is_subsequence(s1,s2))
# time complexity: O(m+n)
```

```python
# with recursion
def is_subsequence(s1, s2, m, n):
    # m length of s1
    # n length of s2

    if n==0:
        return True
    if m==0:
        return False
    if s1[m-1]==s2[n-1]:
        return is_subsequence(s1,s2, m-1, n-1)
    else:
        return is_subsequence(s1, s2, m-1, n)
    
print(is_subsequence(s1='abcd',s2='acd',m=4,n=3))
# time complexity: O(max(m,n))
```
