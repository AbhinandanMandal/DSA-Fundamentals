```python
# Generating subsequence
def subsequence(s:str, curr="", index=0):
    if index == len(s):
        print(curr)
        return

    # Include current character
    subsequence(s, curr+s[index], index+1)
    # Exclude current character
    subsequence(s, curr, index+1)

subsequence("abc")

# Time complexity: O(n.2^n)
# Space complexity: O(n.2^n)
```

```python
def is_subsequence(s1:str, s2:str):
    i=0
    j=0

    while (i<len(s1) and j<len(s2)):
        if s1[i]==s2[j]:
            i+=1
            j+=1
        else:
            j+=1
    
    if j==len(s2):
        return True
    else:
        return False
    
print(is_subsequence('ABCDE', 'ADE'))

# Time complexity: O(m+n)
# Space complexity: O(1)
```

```python
def subsequence(s1:str, s2:str, m:int, n:int):
    if n==0:
        return True
    if m==0:
        return False
    
    if s1[m-1] == s2[n-1]:
        return subsequence(s1, s2, m-1, n-1)
    else:
        return subsequence(s1, s2, m-1, n)
    
print(subsequence(s1='abcd', s2='acd',m=4,n=3))

# Time complexity: O(max(m,n))
# Space complexity: O(max(m,n))
```
