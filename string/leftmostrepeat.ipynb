```python
def leftmost_repetating_character(s:str):
    hash_s={}
    list_s=[]

    # hash formation
    for char in s:
        hash_s[char] = hash_s.get(char, 0)+1
    
    # list formation
    for k,v in hash_s.items():
        if v>=2:
            list_s.append(k)

    # main checking condition
    if len(list_s)>=1:
        idx = s.find(list_s[0])
        return idx 
    return -1


s = 'geeksforgeeks'
print(leftmost_repetating_character(s=s))

# Time complexity: O(n)
# Space complexity: O(c)
```

```python
# mehtod 2
def leftmost_repeat(s:str):
    count=[0]*256
    
    for i in range(len(s)):
        count[ord(s[i])]+=1 # order storing increment by 1 for each character
    
    for i in range(len(s)):
        if count[ord(s[i])]>1: # checking if count of character > 1
            return i
    
    return -1

# s='geeksforgeeks'
s='abcd'
print(leftmost_repeat(s=s))
# time complexity: O(n)
```

```python
# method 3

import sys
def leftmost(s:str):
    findex=[-1]*256
    res=sys.maxsize

    for i in range(len(s)):
        if (findex[ord(s[i])]==-1):
            findex[ord(s[i])]=i
        else:
            res=min(res,findex[ord(s[i])]) # checking index and result

    if res==sys.maxsize:
        return -1
    return res


s='abba'
print(leftmost(s=s))

# time complexity: O(n)
```

```python
# method 4
# first repeating character index
def leftmost(s:str):
    visited=[False]*256
    res=-1

    for i in range(len(s)-1,-1,-1):
        if visited[ord(s[i])]==True:
            res=i
        else:
            visited[ord(s[i])]=True

    return res


s = 'abba'
print(leftmost(s=s))
# time complexity: O(n)
```

```python
# method 5
# last repeating character index
def leftmost(s: str):
    visited = [False]*256
    res = -1

    for i in range(len(s)):
        if visited[ord(s[i])] == True:
            res = i
        else:
            visited[ord(s[i])] = True

    return res


s = 'abba'
print(leftmost(s=s))
```
