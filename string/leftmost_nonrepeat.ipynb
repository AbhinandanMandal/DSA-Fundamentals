```python
# finding out the index of left most non repeating character
```

```python
def leftmost_repeat(s:str):
    count=[0]*256
    for ch in s:
        count[ord(ch)]+=1
    for ch in s:
        if count[ord(ch)]>1:
            return ch # for leftmost repeating character, it can also be extended for index
    return -1


s='geeksforgeeks'
# s = 'abcd'
print(leftmost_repeat(s=s))
```

```python
def leftmost_nonrepeat(s:str):
    count=[0]*256
    for ch in s:
        count[ord(ch)]+=1
    for i in range(len(s)):
        if count[ord(s[i])]==1:
            return i
    return -1


s = 'geeksforgeeks'
# s = 'abcd'
print(leftmost_nonrepeat(s=s))
```
