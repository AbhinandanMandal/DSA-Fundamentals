```python
def leftmost_repeat(s:str):
    count=[0]*256

    for char in s:
        count[ord(char)]+=1
    for char in s:
        if count[ord(char)]>1:
            return char 
    return -1


print(leftmost_repeat('geeksforgeeks'))
print(leftmost_repeat('abcd'))
```

```python
def leftmost_nonrepeat(s:str):
    count=[0]*256

    for char in s:
        count[ord(char)]+=1
    
    for char in s:
        if count[ord(char)]==1:
            return char 
    return -1 


print(leftmost_nonrepeat('geeksforgeeks'))
print(leftmost_nonrepeat('abcd'))
```
