```python
def num_paliendrome(num:int):
    rev=0
    i=0
    temp=num
    while num>0:
        ld=num%10
        rev=rev*10+ld
        num=num//10
        i+=1
    return rev==temp

print(num_paliendrome(num=1001))
```

```python
def str_paliendrome(s:str):
    return s==s[::-1]

print(str_paliendrome(s='abba'))
```

```python
# using two pointer approach
def isPaliendrome(s:str):
    i=0
    j=len(s)-1
    while(i<=j):
        if s[i]!=s[j]:
            return False
        i+=1
        j-=1
    return True

s='ABBA'
print(isPaliendrome(s=s))
```
