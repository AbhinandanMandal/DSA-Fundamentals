A panagram is a string that contains all the 26 characters (a-z) at least one, regarless of case

```python
def is_panagram(s:str):
    seen=set()
    s=s.lower()

    for char in s:
        if 'a'<= char <= 'z':
            seen.add(char)
    
    return len(seen) == 26


s = "Thequickbrownfoxjumpsoverthelazydog"
print(is_panagram(s=s))

# Time complexity: O(n)
# Space complexity: O(1)
```

```python
# missing character in panagram
# for ascii ord('a')=97

def missing_panagram(s:str):
    s=s.lower()
    marked=[False]*26

    for ch in s:
        if 'a'<=ch<='z':
            marked[ord(ch)-ord('a')]=True # marking the present character

    missing=[]
    for i in range(26):
        if not marked[i]: # if marked[i] = False
            missing.append(chr(i+ord('a'))) # accumulating the missing

    return "".join(missing)

    
s = 'Abcdefghijklmnopqrstuvwxy'
print(missing_panagram(s=s))
```
