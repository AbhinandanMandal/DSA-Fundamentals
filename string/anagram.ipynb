```python
# checking anagram
# in anagram, the frquency of characters in both the string must be same
# order does not matter in anagram


# method 1
def anagram(s1:str,s2:str):
    s1_hash={}
    s2_hash={}

    for char in s1:
        s1_hash[char]=s1_hash.get(char,0)+1

    for char in s2:
        s2_hash[char]=s2_hash.get(char, 0)+1

    for k,v in s1_hash.items():
        if k in s2_hash.keys():
            if v!=s2_hash[k]:
                return False
    return True


s1='aab'
s2='abb'
print(anagram(s1=s1,s2=s2))
# time complexity: O(n)
```

```python
# method 2
def anagram(s1,s2):
    if len(s1)!=len(s2):
        return False
    s1=sorted(s1)
    s2=sorted(s2)
    return (s1==s2)


s1 = 'acb'
s2 = 'abc'
print(anagram(s1=s1, s2=s2))
# time complexity: O(nlogn)
```

```python
# method 2

def anagram(s1, s2):
    if len(s1)!=len(s2):
        return False
    count=[0]*256
    for i in range(len(s1)):
        count[ord(s1[i])]+=1
        count[ord(s2[i])]-=1
    
    # if a string is anagram then at end
    # ord must have to be 0 valued

    for x in count:
        if x!=0:
            return False
    return True


s1 = 'abc'
s2 = 'acb'
print(anagram(s1=s1, s2=s2))
# time complexity: O(n)
```
