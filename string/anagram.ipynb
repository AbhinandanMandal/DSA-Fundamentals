```python

def anagram(s1,s2):
    if len(s1)!=len(s2):
        return False
    s1=sorted(s1)
    s2=sorted(s2)
    return (s1==s2)


s1 = 'acb'
s2 = 'abc'
print(anagram(s1=s1, s2=s2))

# Time complexity: O(nlogn)
# Space complexity: O(n)
```
