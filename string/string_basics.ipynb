```python
# char to order and vice versa
print(ord('a'))
print(ord('A'))

print(chr(97))
print(chr(65))

# strings are immutable
s='abhinandan'
# s[0]='A' # throws an error
# print(s)
```

```python
# escape sequence
print("hey what\'s up abhi")
print("hi\nwelcome to the jungle")
print('hey\\')
print('\\n')
```

```python
# getting index of string inside a string
s1='geeksforgeeks'
s2='geeks'

print(s1.index(s2)) # returns the initial position
print(s1.rindex(s2)) # returns the final position
print(s1.index(s2,0,13)) # search from 1st to last position and returns the index of index
s2=s1.upper()
print(s2)
print(s2.isupper())
print(s2.islower())

s3='gfg'
# print(s1.index(s3)) # returns an error
```

```python
s1='abhi is doing dsa course'
print(s1.startswith('abhi'))
print(s1.endswith('course'))
print(s1.startswith('abhi',0,len(s1)))
print(s1.endswith('course',0,len(s1)))
```

```python
s1='abhi is doing dsa'
print(s1.split()) # returns the list
s2 = 'abhi, is, doing, dsa'
print(s2.split(",")), # splitting by not taking commas (,)

l=['abhi','coding','pydsa']
print(' '.join(l))
print(', '.join(l))
```

```python
# strip is used to remove unwanted string character from the beginning and ending of string
s='--abhinandan---'
print(s.strip('-')) # remvoing from both sides
print(s.lstrip('-')) # removing from front
print(s.rstrip('-')) # removing from back
```

```python
# find method gives the starting index of any string inside another string
s1='geeksforgeeks'
s2='geeks'
print(s1.find(s2))
print(s1.find('gfg')) # returns -1 if not present
print(s1.find(s2,0,len(s1)))

# similarly works like .index() method
# .find() dosen't return value error like .index() method
```
