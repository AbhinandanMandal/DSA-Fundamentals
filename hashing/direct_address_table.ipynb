**Hashing**: hashing performs *search, insert, delete* in O(1) time

```python
# Imagning hash table as dictionary
class HashTable:
    def __init__(self, size:int):
        self.size=size
        self.hash_table={}

        for i in range(size):
            self.hash_table[i+1]=i # A hash table formation

    def insert(self, key:int, value:int):
        self.hash_table[key]=value
        return self.hash_table
    
    def search(self, value:int):
        # for i in range(self.size):
        #     if self.hash_table[i]==value:
        #         return True
        # return False 

        # short method
        for v in self.hash_table.values():
            if v==value:
                return True
        return False

    def delete(self, value:int):
        for k,v in self.hash_table.items():
            if v==value:
                del self.hash_table[k] # delete that number
                return f"{value} deleted"
        return f"{value} not found"
            
hash_table=HashTable(100)
print(hash_table.insert(50,12345)) 
print(hash_table.search(12345))
print(hash_table.delete(12345))
print(hash_table.search(12345))
```

When we know keys in advance, then we'll use hashing else use chaining, open addressing

In **Collision Handelling** in hash, we have two methods

1. **Chaining** [[]]: chaining can be implemented with `linked list, dynamic size arrays, self balancing BST`
2. **Open Addressing** (linear probing, quadratic probing, double hashing) []: open addressing handel collision with a single array only. For open addressing, `no of slots in hash table >= no of keys to be inserted`.
