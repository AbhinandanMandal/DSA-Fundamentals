**open addressing**

open addressing handel collision with a single array only. Its catche friendly

for open addressing, `no of slots in hash table >= no of keys to be inserted`

`linear probing, quadratic probing, double hashing`

```python
def LinearProbing(arr:list, n:int):
    hash_table=[-1]*n # n total bucket length
    for num in arr:
        index=num%n # index is hash key
        if hash_table[index]==-1:
            hash_table[index] = num 
            continue
        if hash_table[index]==num: # dublicate element
            continue

        else:
            for i in range(1,n):
                k=(index + i)%n # handel clustring

                if hash_table[k]==-1:
                    hash_table[k]=num
                    break

                if hash_table[k]==num:
                    break
    return hash_table


arr = [4, 14, 24, 44]
n = 10
print(LinearProbing(arr, n))
```

```python
# Same as Linear Probing but with a little twist in cluster handelling
def quadraticProbing(arr:list, m:int):
    hash_t=[-1]*m
    for num in arr:
        index=num%m 
        if hash_t[index]==-1:
            hash_t[index]=num
            continue
        
        if hash_t[index]==num:
            continue
        else:
            for i in range(1,m):
                k=(index+i*i)%m # handel clustering 
                if (hash_t[k]==-1):
                    hash_t[k]=num
                    break
                if (hash_t[k]==num):
                    break
    return hash_t


arr = [21, 10, 33, 43]
m = 5
print(quadraticProbing(arr=arr, m=m))
```
