Collision handelling has 2 methods, chaining and open addressing.

Chaining can be implemented with linked list, dynamic sized array and binary search tree.

```python
class MyHash:
    def __init__(self, b): # b is the key
        self.BUCKET=b
        self.hash = [[] for i in range(b)] # creating like [[], [], ...]

    def insert(self, n:int):
        hash_key= n % self.BUCKET
        self.hash[hash_key].append(n)
        return self.hash 

    def search(self, n:int):
        hash_key= n% self.BUCKET
        if n in self.hash[hash_key]:
            return True
        return False 
    
    def delete(self, n:int):
        hash_key= n%self.BUCKET
        if n in self.hash[hash_key]:
            self.hash[hash_key].remove(n)
            return self.hash 
        else:
            return -1
        
hash = MyHash(7)
print(hash.insert(70))
print(hash.insert(71))
print(hash.insert(9))
print(hash.insert(56))
print(hash.insert(72))
print(hash.search(70))
print(hash.delete(70))
print(hash.search(70))

# normally
# time complexity: O(1), wrost case O(n)
# space complexity: O(n+b), n is number to be inserted, b is the number of bucket in hashtable
```
