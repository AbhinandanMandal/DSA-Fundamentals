```python
# Index of the last element that is greater than or equal to a given number
def _last_element_index(arr:list, n:int):
    low=0
    high=len(arr)-1
    res=-1
    while(low<=high):
        mid=low+(high-low)//2
        if (arr[mid]>=n):
            res=mid
            low=mid+1 # moving to right direction
        else:
            high = mid-1 # moving to left direction
    return res


nums = [1, 2, 8, 10, 10, 12, 19]
print(_last_element_index(nums, 10))

# Time complexity: O(logn)
# Space complexity: O(1)
```
