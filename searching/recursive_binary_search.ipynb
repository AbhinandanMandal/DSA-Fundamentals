```python
def recursive_binary_search(nums:list, n:int, high:int, low:int):
    if(low>=high):
        return -1
    mid=(low+high)//2
    if(nums[mid]==n):
        return mid
    elif(nums[mid]>n):
        high=mid-1
        return recursive_binary_search(nums=nums, n=n, high=high, low=low)
    else:
        low=mid+1
        return recursive_binary_search(nums=nums, n=n, high=high, low=low)
    

nums = [10, 20, 30, 40, 50, 60]
high = len(nums)
# Time complexity: O(logn)
# Space complexity: O(logn)
```
