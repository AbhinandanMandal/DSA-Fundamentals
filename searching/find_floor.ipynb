```python
# Finding the last element that is less than or equal to given number of list
def _last_emenet_list(arr:list, n:int):
    low=0
    high=len(arr)-1
    res=-1

    while(low<=high):
        mid=low+(high-low)//2
        if (arr[mid]<=n):
            res=mid
            low=mid+1
        else:
            high=mid-1
    return res 


nums = [1, 2, 8, 10, 10, 12, 19]
print(_last_emenet_list(nums, 10))

# Time complexity: O(logn)
# Space complexity: O(1)
```
