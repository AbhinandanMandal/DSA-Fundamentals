```python
# Finding index of a number
def linear_search(nums:list, n:int):
    for i in range(len(nums)):
        if(nums[i]==n):
            return i
    return -1


nums = [0, 20, 80, 30, 60, 50, 110, 100, 130, 170]
print(linear_search(nums=nums, n=30))

# Time complexity: O(n)
# Space complexity: O(1)
```

```python
# using two pointer approach
def linear_search(nums:list, n:int):
    left=0
    right=len(nums)-1

    while left<=right:
        if(nums[left]==n):
            return left
        if (nums[right]==n):
            return right
        left+=1
        right-=1

    return -1


nums = [10, 20, 30, 40, 50]
print(linear_search(nums=nums, n=20))

# Time complexity: O(n)
# Space complexity: O(1)
```

```python
def _binary_search(arr:list, n:int):
    low=0
    high=len(arr)-1
    while(low<=high):
        mid=(low+high)//2
        if (arr[mid]==n):
            return mid
        elif (arr[mid]>n):
            high=high-1
        else:
            low+=1
    return -1


nums = [10, 20, 30, 40, 50]
print(_binary_search(nums, 20))

# Time complexity: O(logn)
# Space complexity: O(1)
```
