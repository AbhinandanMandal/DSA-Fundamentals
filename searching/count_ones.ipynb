```python

def first_occurance(nums:list):
    low=0
    high=len(nums)-1
    while(low<=high):
        mid=(low+high)//2
        if(nums[mid]>1):
            high=mid-1
        elif(nums[mid]<1):
            low=mid+1
        else:
            if(mid==0 or nums[mid-1]!=nums[mid]):
                return mid
            else:
                high=mid-1
    return -1

def last_occurance(nums:list):
    low=0
    high=len(nums)-1
    while (low <= high):
        mid = (low+high)//2
        if (nums[mid] > 1):
            high = mid-1
        elif (nums[mid] < 1):
            low = mid+1
        else:
            if (mid == len(nums)-1 or nums[mid] != nums[mid+1]):
                return mid
            else:
                low=mid+1
    return -1


def count_occurance(nums:list):
    if(first_occurance(nums=nums)==-1):
        return 0
    else:
        return last_occurance(nums=nums)-first_occurance(nums=nums)+1
    

nums = [0, 0, 0, 1, 1, 1, 1]
print(count_occurance(nums=nums))  
# Time complexity: O(logn)
# Space complexity: O(1)
```

```python
# condition: array must be sorted in non-decreasing order
def _count_ones(nums:list):
    low=0
    high=len(nums)-1
    while(low<=high):
        mid=(low+high)//2
        if(nums[mid]==1 and (mid==len(nums)-1 or nums[mid+1]==0)): 
            # If current element is 1 and mid elemnt is last of the array or next mid+1=0 then return mid+1
            return mid+1
        elif(nums[mid]==0):
            high=mid-1
        elif(nums[mid]==1):
            low=mid+1
    return -1


nums = [1, 1, 1, 1, 0, 0, 0]
print(_count_ones(nums=nums))  
# Time complexity: O(logn)
# Space complexity: O(1)
```
