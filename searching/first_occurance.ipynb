```python
def _first_occurance(arr:list, n:int):
    low=0 
    high=len(arr)-1
    while(low<=high):
        mid=(low+high)//2
        if (arr[mid]>n):
            high=mid-1
        elif (arr[mid]<n):
            low=mid+1
        else:
            if (mid==0 or arr[mid-1]!=arr[mid]):
                return mid
            else:
                high=mid-1
    return -1


nums = [1, 2, 2, 2, 3, 3, 3, 4]
print(_first_occurance(nums=nums, n=2))  
# Time complexity: O(logn)
# Space complexity: O(1)
```
