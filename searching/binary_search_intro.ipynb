Binary Search works well for sorted array

```python
# Also known as linear search
def _finding_element(arr:list, n:int):
    for i in range(len(arr)):
        if arr[i]==n:
            return i
    return -1


arr = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
print(_finding_element(arr, 5))
# Time complexity: O(n)
# Space complexity: O(1)
```

```python
def _binary_search(arr, n):
    low=0
    high=len(arr)-1
    while (low<=high):
        mid=(low+high)//2
        if arr[mid]==n:
            return mid 
        if arr[mid]>n:
            high=mid-1
        else:
            low=mid+1
    return -1


arr = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
print(_binary_search(arr, 5))
# Time complexity: O(logn)
# Space complexity: O(1)
```
